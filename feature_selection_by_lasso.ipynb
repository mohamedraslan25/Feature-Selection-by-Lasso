```python
X = df.drop('MEDV', axis = 1)
y = df['MEDV']
```

```python
X = X.values
y = y.values
```

```python
from sklearn.linear_model import Lasso
```

```python
alpha = 0.1
lasso = Lasso(alpha = alpha)
```

```python
lasso.fit(X, y)
```

```python
lasso.score(X, y)
```

```python
y_predict = lasso.predict(X)
```

```python
y_predict
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, explained_variance_score, mean_squared_log_error, max_error
```

```python
mse = mean_squared_error(y, y_predict)
rmse = mean_squared_error(y, y_predict, squared=False)
mae = mean_absolute_error(y, y_predict)
acc = r2_score(y, y_predict)
evs = explained_variance_score(y, y_predict)
# msle = mean_squared_log_error(y, y_predict)
max_err = max_error(y, y_predict)
```

```python
print('mean_squared_error: {0} \nroot_mean_squared_error: {1} \nmean_absolute_error: {2} \nr2_score: {3} \nexplained_variance_score: {4} \nmax_error: {5}'.format(mse, rmse, mae, acc, evs, max_err))
```
